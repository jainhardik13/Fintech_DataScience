```
import yfinance as yf
```

```
# Taking input of the stock symbol and then it will fetch on the yahoo finance website and print the first 100 lines on the console
ticker = input("Enter the stock symbol: ")
data = yf.download(ticker, start="2025-01-01", end="2026-02-04")
data.head(100)
```

```
# Daily Closing data of the stock
daily_close = data["Close"]
print(daily_close)
```

```
# plotting a simple graph of daily closing price of a stock
plt.figure(figsize=[12, 8])
plt.plot(daily_close)
plt.title("Daily closing price")
plt.show()
```

```
# Drawing graph using seaborn
daily_close_df = daily_close.reset_index()
daily_close_df.columns = ["Date", "Close"]
plt.figure(figsize=[12, 8])
sns.lineplot(data=daily_close_df, x="Date", y="Close")
plt.title("Daily closing price")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```
# Creating a new column 7 days moving average and 30 days moving average
data["7_day_MA"] = data["Close"].rolling(window=7).mean()
data["30_day_MA"] = data["Close"].rolling(window=30).mean()
print(data)
```

```
plt.figure(figsize=[10, 5])
plt.plot(data["Close"],label="Daily close price")
plt.plot(data["7_day_MA"],label="7-day MA")
plt.plot(data["30_day_MA"],label="30-day MA")
plt.title(f" {ticker} Stock price time series")
plt.legend()
plt.tight_layout()
plt.show()
```
